```python
def leer_archivo_y_generar_listas(nombre_archivo):
    """
    Lee un archivo de texto con dos strings separados por comas y genera dos listas.

    Args:
        nombre_archivo (str): La ruta del archivo de texto.

    Returns:
        tuple: Dos listas con los datos del archivo.
    """

    try:
        # Leer el archivo de texto con Pandas
        df = pd.read_csv(nombre_archivo, header=None, names=['Asignatura1', 'Asignatura2'])

        # Convertir las columnas en listas
        lista1 = df['columna1'].tolist()
        lista2 = df['columna2'].tolist()

        return lista1, lista2

    except FileNotFoundError:
        print(f"Error: El archivo '{nombre_archivo}' no fue encontrado.")
        return [], []
    except Exception as e:
        print(f"Ocurrió un error inesperado: {e}")
        return [], []

```

```python
# Ejemplo de uso
nombre_archivo = "asignaturas.txt"  # Reemplaza con el nombre de tu archivo

#Generar un fichero de ejemplo si no existe.
try:
    with open(nombre_archivo, 'x') as f:
        f.write('valor1,valor2\nvalor3,valor4\nvalor5,valor6')
except FileExistsError:
    print('El fichero ya existe')

lista1, lista2 = leer_archivo_y_generar_listas(nombre_archivo)

if lista1 and lista2:
    print("Lista 1:", lista1)
    print("Lista 2:", lista2)
```
