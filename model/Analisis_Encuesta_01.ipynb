# Analisis de los datos de la encuesta Jumping Girls

```python
import pandas as pd
```

```python
encuesta_ESP = "Encuesta Digital Girls_SP&ENG_A.xlsx"
```

```python
df = pd.read_excel(encuesta_ESP)
df.head()
```

```python
print("Tenemos ", df['Label'].count(), "respuestas de la encuesta en Español")
```

```python
# por categorias profesionales
df['Label'].value_counts()
```

```python
encuesta_ENG = 'Encuesta Digital Girls_SoloENG_A.xlsx'
df_ENG = pd.read_excel(encuesta_ENG)
df_ENG.head()
```

```python
print("Tenemos ", df_ENG['Label'].count(), "respuestas de la encuesta en Inglés")
```

```python
# por categorias profesionales
df_ENG['Label'].value_counts()
```
